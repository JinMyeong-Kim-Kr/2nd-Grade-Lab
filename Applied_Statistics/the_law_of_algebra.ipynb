## 1. 대수의 법칙 ##

```python
# (1) 주사위 던지는 함수 구현, 10번 던진 결과값을 print
import random

def throw_dice(num):
  vals = []
  for i in range(num):
    vals.append(random.randrange(1,7))
  return vals
    
print(throw_dice(num=10))
```

```python
# (2) 주사위를 여러번 던졌을 때, 눈들의 평균값 및 표준편차의 추이
import numpy
import matplotlib.pyplot as plt

avg = []
s_deviation = []
num_max = 500
nums = list(range(10, num_max))
# << TODO >>

for i in nums:
  tmp = throw_dice(i)
  avg.append(numpy.mean(tmp))
  s_deviation.append(numpy.std(tmp))
plt.plot(avg)
plt.plot(s_deviation)
plt.show()
```

```python
# (3) 주사위를 1000번, 10000번 던졌을 때는?
avg = []
s_deviation = []
num_max = 1000
nums = list(range(10, num_max, 1))

for i in nums:
  tmp = throw_dice(i)
  avg.append(numpy.mean(tmp))
  s_deviation.append(numpy.std(tmp))
plt.plot(avg)
plt.plot(s_deviation)
plt.show()

num_max = 10000
nums = list(range(10, num_max, 10))
avg = []
s_deviation = []

for i in nums:
  tmp = throw_dice(i)
  avg.append(numpy.mean(tmp))
  s_deviation.append(numpy.std(tmp))
plt.plot(nums,avg)
plt.plot(nums,s_deviation)
plt.show()
```
