## 2. 원주율 구하기 ##

```python
# (1) x,y~[-1,1] 평면상에 임의의 점을 생성하는 함수 구현, scatter 함수로 시각화
import random
import matplotlib.pyplot as plt

def xy_rand(num):
  x = []
  y = []
  for i in range(num):
    x.append(random.uniform(-1,1))
    y.append(random.uniform(-1,1))
  return x,y

num = 100
x, y = xy_rand(num)
plt.scatter(x, y)
plt.show()
```

```python
# (2) 생성된 x,y 점들에 대해서 개별 점들 중, 단위 원 x^2+y^2=1 안에 있는 점들의 비율을 계산 후, 원주율을 근사
import numpy as np
num = 100
x, y = xy_rand(num)
cnt = 0
tmp = np.sqrt(np.square(x) + np.square(y)) < 1
pi_approx = np.sum(tmp)/num * 4
print(pi_approx)
```

```python
# (3) 생성된 점의 개수를 100개에서 10000개, 20000개를 늘렸을 때 원주율의 값은?
num = 10000
x, y = xy_rand(num)
tmp = np.sqrt(np.square(x) + np.square(y)) < 1
pi_approx = np.sum(tmp)/num * 4
print(pi_approx)

num = 20000
x, y = xy_rand(num)
tmp = np.sqrt(np.square(x) + np.square(y)) < 1
pi_approx = np.sum(tmp)/num * 4
print(pi_approx)
```

```python
# (4) range(1000, 500000, 1000)에 대해서 추정된 원주율 값의 추이를 그리면?, 실제 pi값 (np.pi)를 그래프에 점선으로 그려서 표시
import numpy as np

nums = list(range(1000, 500000, 2000))
cnt = 0
pi_x = []
for k in nums:
  x, y = xy_rand(k)
  tmp = np.sqrt(np.square(x) + np.square(y)) < 1
  pi_approx = np.sum(tmp)/k * 4
  pi_x.append(pi_approx)
    
plt.plot(nums,pi_x) 
plt.axhline(y=np.pi, xmin=0, xmax=250,linestyle='dashed',color='orange')
plt.show()
```
