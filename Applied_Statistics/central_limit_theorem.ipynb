## 1. 중심극한정리 (central limit theorem)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
# (1) 이항분포 확률분포 (N=10, p=0.2) 객체 정의 및 확률분포함수를 plot

binom = stats.binom
n = 10
p = 0.2
x = np.arange(0,11)
plt.plot(x, binom.pmf(x, n, p))
```

```python
# (2) 이항분포 확률분포의 객체와 표본 크기 N_samples를 입력변수로 받아, 표본평균을 생성하는 함수를 구현
# rvs함수의 size 인자를 이용할것

def generate_sample_mean(distrib, N_samples=50):
    ## TODO ##
    rvs = distrib.rvs(10, 0.2, size = N_samples)
    return np.mean(rvs)
    
print(generate_sample_mean(binom, 50))
```

```python
# (3) 표본평균을 10000개 생성하여 히스토그램을 (bin개수 50개) 표시. N_samples를 10개, 50개, 500개로 시도

tmp = []
for i in range(10000):
      tmp.append(generate_sample_mean(binom, N_samples=10))

x_bars = np.array(tmp)
plt.figure(0)
plt.hist(x_bars, bins=50)
plt.show()
print('N_samples=10', (np.mean(x_bars), np.std(x_bars)))

tmp = []
for i in range(10000):
      tmp.append(generate_sample_mean(binom, N_samples=50))
x_bars = np.array(tmp)

plt.figure(0)
plt.hist(x_bars, bins=50)
plt.show()
print('N_samples=50', (np.mean(x_bars), np.std(x_bars)))

tmp = []
for i in range(10000):
      tmp.append(generate_sample_mean(binom, N_samples=500))
x_bars = np.array(tmp)
plt.figure(0)
plt.hist(x_bars, bins=50)
plt.show()
print('N_samples=500', (np.mean(x_bars), np.std(x_bars)))
```

```python
# (4) 다른 확률분포 (stats.distributions) 에서 생성된 표본평균의 분포도 볼 수 있다.

## TODO ##
chi2 = stats.chi2
df = 10
x = np.linspace(0, 10)
plt.plot(x, chi2.pdf(x, df), label='chi2 pdf')
plt.show()

tmp = []
for i in range(10000):
      tmp.append(generate_sample_mean(chi2, N_samples=50))
x_bars = np.array(tmp)
plt.hist(x_bars, bins=50)
plt.show()
print('N_samples=50', (np.mean(x_bars), np.std(x_bars)))
```
